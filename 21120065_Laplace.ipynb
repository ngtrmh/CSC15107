<div style='font-size: 16px;'>
<strong id="3.2.2a">▶ Tại sao cần dùng cơ chế Laplace thay vì cho người dùng phản hồi tự chọn tùy ý một giá trị nhiễu?</strong>
<ol>
    <li>Đảm bảo tính ngẫu nhiên và phân phối có kiểm soát</li>
    Khi sử dụng cơ chế Laplace, nhiễu được sinh ra từ một phân phối có dạng chuẩn với tham số $b$ (scale parameter). Phân phối này có đặc điểm:
    <br>- Kỳ vọng (mean) của phân phối là 0, nghĩa là trung bình nhiễu được thêm vào sẽ không làm thay đổi giá trị trung bình thực tế của dữ liệu.
    <br>- Phân phối Laplace có xác suất cao để sinh ra những giá trị nhiễu nhỏ và xác suất thấp hơn để sinh ra những giá trị nhiễu lớn, tạo ra một mức độ bảo vệ đồng đều.
    </p>
    <li>Tính toán được mức độ riêng tư ($\epsilon$)</li>
    <p>Cơ chế Laplace cho phép tính toán và điều chỉnh mức độ riêng tư ($\epsilon$) một cách chính xác. Điều này đảm bảo rằng quyền riêng tư của người tham gia được bảo vệ ở mức độ mong muốn.</p>
    <li>Độ nhạy của hàm truy vấn</li>
    <p> Cơ chế Laplace điều chỉnh mức độ nhiễu dựa trên độ nhạy của hàm truy vấn. Điều này đảm bảo rằng nhiễu thêm vào tương ứng với mức độ ảnh hưởng của từng bản ghi cá nhân trong tập dữ liệu, giúp bảo vệ quyền riêng tư một cách hợp lý.</p>
    <li>Ngăn chặn việc suy diễn ngược dữ liệu cá nhân</li>
    <p>Khi nhiễu được sinh ra ngẫu nhiên từ phân phối Laplace, kẻ tấn công không thể dễ dàng suy đoán hoặc suy diễn ngược dữ liệu cá nhân. Nếu người phản hồi tự chọn nhiễu, có thể sẽ có các mẫu nhiễu không ngẫu nhiên hoặc dễ đoán, làm giảm hiệu quả của việc bảo vệ quyền riêng tư.</p>
    <li>Đảm bảo độ chính xác thống kê của dữ liệu sau gây nhiễu</li>
    <p>Nhiễu từ phân phối Laplace đảm bảo rằng các giá trị thống kê (như tổng, trung bình) của dữ liệu sau khi thêm nhiễu vẫn phản ánh chính xác các đặc điểm của dữ liệu gốc với một mức sai lệch có thể chấp nhận được. Nếu người phản hồi tự chọn nhiễu, các giá trị thống kê có thể bị méo mó nhiều hơn và không phản ánh chính xác dữ liệu thực tế.</p>
</ol>
</div>

<div style='font-size: 16px;'>
    <strong id="3.2.2b">▶ Ảnh hưởng của giá trị tham số $b$</strong>
    <p>Tham số $b$ (scale parameter) trong phân phối Laplace ảnh hưởng trực tiếp đến mức độ nhiễu được thêm vào dữ liệu, từ đó ảnh hưởng đến các giá trị thống kê như tổng và trung bình cộng. Khảo sát dưới đây sẽ cho thấy ảnh hưởng của các giá trị $b$ khác nhau đến các kết quả thống kê.</p>
    <u><i>Dữ liệu thực tế</i></u>
    <p>Giả sử ta có dữ liệu thực tế: $(3, 4, 7, 5, 0)$</p>
    <u><i>Các bước khảo sát</i></u>
    <ol>
        <li>Chọn các giá trị b khác nhau trong khoảng từ $0.01$ đến $2$</li>
        <li>Thêm nhiễu và dữ liệu thực tế</li>
        <li>Tương ứng với mỗi giá trị $b$, tính độ lệch giữa tổng thực tế và tổng sau khi gây nhiễu, độ lệch giữa trung bình cộng thực tế và trung bình cộng sau khi gây nhiễu.</li>
    </ol>
    <u><i>Thực hiện khảo sát</i></u>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu thực tế
data = [3, 4, 7, 5, 0]

# Các giá trị của tham số b
b_values = np.linspace(0.01, 2, 100)

# Số lần chạy để lấy trung bình kết quả
num_runs = 1000

# Hàm thêm nhiễu và tính toán độ lệch
def calculate_deviation(data, b, num_runs):
    total_deviations = []
    average_deviations = []
    
    for _ in range(num_runs):
        noisy_data = [x + np.random.laplace(0, b) for x in data]
        total_noisy = sum(noisy_data)
        average_noisy = total_noisy / len(data)
        
        total_deviation = abs(total_noisy - sum(data))
        average_deviation = abs(average_noisy - (sum(data) / len(data)))
        
        total_deviations.append(total_deviation)
        average_deviations.append(average_deviation)
    
    avg_total_deviation = np.mean(total_deviations)
    avg_average_deviation = np.mean(average_deviations)
    return avg_total_deviation, avg_average_deviation

# Khảo sát các giá trị b khác nhau
total_deviations = []
average_deviations = []

for b in b_values:
    avg_total_dev, avg_avg_dev = calculate_deviation(data, b, num_runs)
    total_deviations.append(avg_total_dev)
    average_deviations.append(avg_avg_dev)

# Vẽ biểu đồ
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(b_values, total_deviations, marker='o')
plt.title('Độ lệch tổng theo b')
plt.xlabel('Tham số b')
plt.ylabel('Độ lệch tổng')

plt.subplot(1, 2, 2)
plt.plot(b_values, average_deviations, marker='o')
plt.title('Độ lệch trung bình theo b')
plt.xlabel('Tham số b')
plt.ylabel('Độ lệch trung bình')

plt.tight_layout()
plt.show()
```

<div style='font-size: 16px;'>
    <p>
    <u><i>Kết quả</i></u>
    <br><i>- Biểu đồ độ lệch tổng: </i>Hiển thị độ lệch trung bình của tổng giữa dữ liệu thực tế và dữ liệu sau khi thêm nhiễu theo các giá trị $b$.
    <br><i>- Biểu đồ độ lệch trung bình: </i>Hiển thị độ lệch trung bình của giá trị trung bình cộng giữa dữ liệu thực tế và dữ liệu sau khi thêm nhiễu theo các giá trị $b$.
    </p>
    <p>
    <u><i>Phân tích kết quả</i></u>
    <br><i>- Khi giá trị $b$ nhỏ, độ lệch tổng và độ lệch trung bình sẽ thấp hơn, nghĩa là nhiễu ít và dữ liệu sau khi thêm nhiễu gần với dữ liệu thực tế hơn.
    <br><i>- Khi giá trị $b$ tăng, độ lệch tổng và độ lệch trung bình sẽ tăng, nghĩa là nhiễu nhiều hơn và dữ liệu sau khi thêm nhiễu xa hơn dữ liệu thực tế, nhưng bảo vệ quyền riêng tư tốt hơn.
    </p>
    <p>
        <u><i>Kết luận</i></u> 
        <br>- Việc lựa chọn giá trị $b$ phù hợp cần phải cân nhắc giữa độ chính xác của dữ liệu và mức độ bảo vệ quyền riêng tư muốn đạt được.
    </p>

<div style='font-size: 16px;'>
    <strong id="3.2.2c">▶ Tính ngẫu nhiên của phát sinh nhiễu</strong>
    <p>Khi thêm nhiễu từ phân phối Laplace, dữ liệu bị nhiễu sẽ có sự khác nhau giữa các lần thêm nhiễu do tính ngẫu nhiên của việc sinh nhiễu. Tuy nhiên, việc sử dụng cơ chế Laplace vẫn đảm bảo sự khác biệt không quá lớn và các giá trị thống kê như tổng và trung bình cộng vẫn phản ánh đúng đặc điểm của dữ liệu thực tế trong dài hạn.</p>
    <u><i>Cách giảm thiểu sự khác biệt:</i></u>
    <ol>
        <li><i>Điều chỉnh tham số $b$:</i> Tham số $b$ xác định mức độ nhiễu. Một giá trị $b$ nhỏ sẽ tạo ra nhiễu ít hơn, nhưng cũng có thể làm giảm mức độ bảo vệ quyền riêng tư. Có thể cân nhắc giữa độ bảo mật và độ chính xác bằng cách thử nghiệm với các giá trị $b$ khác nhau.</li>
        <li><i>Tăng số lượng mẫu:</i> Khi có nhiều mẫu hơn, sự khác biệt trong các lần chạy sẽ ít ảnh hưởng đến giá trị thống kê trung bình. Số lượng lớn các giá trị bị nhiễu sẽ làm trung bình các nhiễu và giúp các giá trị thống kê trở nên ổn định hơn.</li>
    </ol>
    <u><i>Ví dụ:</i></u>
    <br><i>- Dữ liệu thực tế:</i> $(3, 4, 7, 5, 0)$
    <br><i>- Chọn tham số:</i> $b = 0.02$
    <br><i>- Thêm nhiễu, tính toán tổng và trung bình trong $1000$ lần chạy</i>
</div>

```python
import numpy as np

# Dữ liệu thực tế
data = [3, 4, 7, 5, 0]

# Tham số của phân phối Laplace
b = 0.02

# Chạy nhiều lần để quan sát sự khác biệt
num_runs = 1000
totals_noisy = []
averages_noisy = []

for _ in range(num_runs):
    noisy_data = [x + np.random.laplace(0, b) for x in data]
    total_noisy = sum(noisy_data)
    average_noisy = total_noisy / len(data)
    totals_noisy.append(total_noisy)
    averages_noisy.append(average_noisy)
    
print("Khoảng giá trị các tổng sau khi gây nhiễu trong các lần chạy: [",min(totals_noisy),", ",max(totals_noisy),"]")
print("Khoảng giá trị các trung bình sau khi gây nhiễu trong các lần chạy: [",min(averages_noisy),", ",max(averages_noisy),"]")
```

<div style='font-size: 16px;'>
    <u><i>Phân tích kết quả</i></u>
    <br><i>- Trung bình của tổng sau khi gây nhiễu: </i>Trung bình của các giá trị tổng bị nhiễu qua 100 lần chạy là gần với tổng thực tế (19). Điều này cho thấy mặc dù có sự khác biệt giữa các lần chạy, nhưng trung bình các giá trị nhiễu vẫn phản ánh đúng đặc điểm của dữ liệu thực tế.
    <br><i>- Trung bình của trung bình cộng sau khi gây nhiễu: </i>Tương tự, trung bình của các giá trị trung bình cộng bị nhiễu cũng gần với giá trị trung bình cộng thực tế (3.8)
